```python
import MySQLdb
import pandas as pd
import matplotlib.pyplot as plt
from textwrap import wrap
from matplotlib.ticker import FuncFormatter



variables = ["age_range","gender","race","marst","educ","employ","livarange","primincome","region","noprior","sub1","freq1","frstuse1","sub2","sub3","numsubs","alcdrug","psyprob"]

conn = MySQLdb.connect(host="localhost", user="snuthakk", passwd="6115096", db="group2_db")
cursor = conn.cursor()

df1 = pd.read_sql_query("select  age_range , gender , race , marst , educ , employ , livarange , primincome , region , noprior , sub1 , freq1 , frstuse1 , sub2 , sub3 , numsubs , alcdrug , psyprob  from admission_records;", conn)

def counter(df):
    dict = {}
    for i in df:
        if i not in dict:
            dict[i] = 1
        else:
            dict[i] += 1
    return dict


for i in variables:
    if i != "sub1" and i!= "sub3" and i!= "race" and i!= "region":
        a = counter(df1[i])
        labels = list(a.keys())
        sizes = list(a.values())
        fig1, ax1 = plt.subplots()
        ax1.pie(sizes, labels=labels,autopct='%1.0f%%')
        ax1.axis('equal')
        plt.title(i)
        plt.show()
    else:
        a = counter(df1[i])
        length = len(a)
        empty = []
        fig, ax = plt.subplots()
        for i in range(1,length+1):
            empty.append(i)
        labels = list(a.keys())
        labels = labels = [ '\n'.join(wrap(l, 15)) for l in labels ]
        sizes = list(a.values())
        plt.bar(empty,sizes,tick_label = labels)
        plt.xticks(rotation="vertical")
        plt.tight_layout()
        plt.show()
```
